# PCA Visualization

```python
#in pca 1st step is to do standard normal distribution
from sklearn.preprocessing import StandardScaler
```

```python
scale=StandardScaler()
scale.fit(df)
```

```python
scaled_data=scale.transform(df)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2) #n_components is that how many dimensions we are going to convert the data
```

```python
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
x_pca.shape
```

```python

scaled_data.shape
```

```python
scaled_data
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer["target"],cmap="plasma")
plt.xlabel("first priciple component")
plt.ylabel("second priciple component")
```

```python
#hyperparameter tuning eith randomized searchcv
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import randint
```

```python
est=RandomForestClassifier()
params={"max_depth"   :[3,4,5,10,None],
        "n_estimators":[100,200,300,400],
        "criterion":["gini","entropy"],
        "bootstrap":[True,False],
        "min_samples_leaf":[1,2,3,4,5]}
def hypertuning_rscv(est,params,nbr_iter,X,y):
    random_search=RandomizedSearchCV(est,params,n_jobs=-1,n_iter=nbr_iter,cv=10)
    random_search.fit(X,y)
    ht_params=random_search.best_params_
    ht_score=random_search.best_score_
    return ht_params,ht_score
```

```python
parameters,score=hypertuning_rscv(est,params,1,X,y)
```
